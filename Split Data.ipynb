```python
import json

input_file = "/Volumes/Personal Drive/GitHub/Yelp_Business_Review_Anlalysis/Yelp JSON/yelp_dataset/yelp_academic_dataset_review.json"

num_files = 10   # Number of files to split into
output_prefix = "split_file_"  # Prefix for output files

# Count lines
with open(input_file, "r", encoding="utf8") as f:
    total_lines = sum(1 for _ in f)

lines_per_file = total_lines // num_files

# Now split into multiple smaller files
with open(input_file, "r", encoding="utf8") as f:

    for i in range(num_files):

        with open(f"{output_prefix}{i+1}.json", "w", encoding="utf8") as out:

            for _ in range(lines_per_file):
                line = f.readline()

                if not line:
                    break

                out.write(line)

print("Split complete.")
```

The Yelp review dataset is around 5GB, which is difficult to load into memory. This script splits the large JSON file into smaller files. First it counts the total number of JSON records by counting lines, then calculates how many lines should go into each split file. It then sequentially reads the original file and writes chunks of lines into multiple smaller files, making it easier to process or upload to systems like Snowflake or S3.

Note: Large JSON datasets are often split before ingestion because systems like Snowflake or Spark parallelize ingestion better when multiple files exist rather than a single large file.
